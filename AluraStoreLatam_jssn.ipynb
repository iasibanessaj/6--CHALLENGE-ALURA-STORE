### Importación de datos



```python
#importación de los datos
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda4.head()
```

#1. Análisis de facturación



```python
ingreso_T_T1 = tienda['Precio'].sum()
ingreso_T_T2 = tienda2['Precio'].sum()
ingreso_T_T3 = tienda3['Precio'].sum()
ingreso_T_T4 = tienda4['Precio'].sum()

print(f'Ingreso total de la tienda 1 es: {ingreso_T_T1} ')
print(f'Ingreso total de la tienda 2 es: {ingreso_T_T2} ')
print(f'Ingreso total de la tienda 3 es: {ingreso_T_T3} ')
print(f'Ingreso total de la tienda 4 es: {ingreso_T_T4} ')
```

# 2. Ventas por categoría



```python
print('--- Cantidad de productos vendidos por categoría en Tienda 1 ---')
ventas_por_categoria_t1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
print(ventas_por_categoria_t1)

print('\n--- Cantidad de productos vendidos por categoría en Tienda 2 ---')
ventas_por_categoria_t2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
print(ventas_por_categoria_t2)

print('\n--- Cantidad de productos vendidos por categoría en Tienda 3 ---')
ventas_por_categoria_t3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
print(ventas_por_categoria_t3)

print('\n--- Cantidad de productos vendidos por categoría en Tienda 4 ---')
ventas_por_categoria_t4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)
print(ventas_por_categoria_t4)
```

```python
print(f'La categoría más popular en Tienda 1 es: {ventas_por_categoria_t1.index[0]} con {ventas_por_categoria_t1.iloc[0]} ventas')
print(f'La categoría más popular en Tienda 2 es: {ventas_por_categoria_t2.index[0]} con {ventas_por_categoria_t2.iloc[0]} ventas')
print(f'La categoría más popular en Tienda 3 es: {ventas_por_categoria_t3.index[0]} con {ventas_por_categoria_t3.iloc[0]} ventas')
print(f'La categoría más popular en Tienda 4 es: {ventas_por_categoria_t4.index[0]} con {ventas_por_categoria_t4.iloc[0]} ventas')
```

# 3. Calificación promedio de la tienda


```python
calificacion_mediat1 = tienda['Calificación'].mean()
calificacion_mediat2 = tienda2['Calificación'].mean()
calificacion_mediat3 = tienda3['Calificación'].mean()
calificacion_mediat4 = tienda['Calificación'].mean()

print(f'Promedio Calificación de la tienda 1 es: {calificacion_mediat1}')
print(f'Promedio Calificación de la tienda 2 es: {calificacion_mediat2}')
print(f'Promedio Calificación de la tienda 3 es: {calificacion_mediat3}')
print(f'Promedio Calificación de la tienda 4 es: {calificacion_mediat4}')
```

## Resumen Final del Análisis Integral

Después de haber realizado un análisis exhaustivo y generado diversas visualizaciones para las cuatro tiendas del Sr. Juan, podemos consolidar los siguientes hallazgos clave:

### 1. Análisis de Facturación (Gráfico de Barras de Ingresos Totales)
*   **Tienda 1** registra los ingresos totales más altos, indicando su posición como la de mayor volumen de ventas.
*   **Tienda 4** consistentemente muestra los ingresos totales más bajos entre las cuatro tiendas, sugiriendo un menor rendimiento financiero general.

### 2. Ventas por Categoría (Gráficos de Barras por Tienda)
*   Las categorías **'Muebles'** y **'Electrónicos'** son universalmente las más vendidas en todas las tiendas, lo que subraya una fuerte demanda en estos segmentos en el mercado general.
*   La **Tienda 3** sobresale en ventas de 'Muebles', mientras que la **Tienda 4** muestra volúmenes notablemente más bajos en 'Electrodomésticos' e 'Instrumentos musicales', áreas donde podría haber oportunidades de mejora o ajuste de inventario.

### 3. Calificación Promedio de la Tienda (Gráfico de Barras de Calificación Promedio)
*   Las calificaciones promedio son muy similares entre todas las tiendas, oscilando entre 3.97 y 4.05, lo que sugiere una satisfacción del cliente generalmente buena.
*   La **Tienda 3** obtiene la calificación promedio ligeramente más alta, mientras que las **Tiendas 1 y 4** comparten las calificaciones más bajas (aunque la diferencia es mínima), indicando posibles áreas para pulir la experiencia del cliente.

### 4. Productos Más Vendidos (Gráficos de Barras de Top 7 Productos)
*   Los productos más vendidos varían por tienda, aunque **'Armario'** (Tienda 1), **'Iniciando en programación'** (Tienda 2 - un punto interesante al ser un libro), **'Kit de bancas'** (Tienda 3), y **'Cama box'** (Tienda 4) destacan. Esto sugiere que las tiendas pueden tener clientelas o enfoques ligeramente diferentes, y que la Tienda 2 tiene un nicho fuerte en contenido educativo.

### 5. Productos Menos Vendidos (Gráficos de Barras de Menos Vendidos)
*   Los productos menos vendidos también difieren, con **'Auriculares con micrófono'** (Tienda 1), **'Juego de mesa'** (Tienda 2), **'Bloques de construcción'** (Tienda 3), y **'Guitarra eléctrica'** (Tienda 4) como ejemplos. Identificar estos productos es crucial para la gestión de inventario y para evitar la acumulación de stock.

### 6. Costo Promedio de Envío (Gráfico de Barras de Costo Promedio de Envío)
*   La **Tienda 4** se distingue por tener el costo promedio de envío más bajo, lo que representa una ventaja operativa y competitiva.
*   La **Tienda 1** presenta el costo de envío más alto, lo cual podría impactar sus márgenes de ganancia o la percepción del cliente si estos costos se transfieren.

### 7. Distribución de Precios General y por Categoría (Histograma y Boxplot)
*   El **histograma** muestra que la mayoría de los productos se concentran en rangos de precios bajos a medios, con una cola extendida hacia precios más altos, indicando la existencia de productos de alto valor con menor frecuencia.
*   Los **gráficos de cajas** revelan una gran variabilidad de precios dentro de categorías como 'Electrodomésticos' e 'Instrumentos musicales', que también presentan valores atípicos (productos de muy alto precio). Categorías como 'Libros' o 'Juguetes' tienen distribuciones de precios más compactas y generalmente más bajas.

### Conclusión General y Recomendación:

Considerando todos los análisis, la **Tienda 4** emerge como la candidata más evidente para una reevaluación estratégica o posible cierre. Aunque posee el costo de envío más competitivo, sus bajos ingresos totales y una de las calificaciones de cliente más bajas la sitúan en una posición de vulnerabilidad en comparación con las otras tiendas. Las visualizaciones confirman que sus debilidades en facturación y satisfacción del cliente son más pronunciadas.
